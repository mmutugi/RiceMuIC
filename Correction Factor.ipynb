```python
#this is the correction factor plotting function which uses the formula (E beam/ E beam-K. Ebeam-B) where B and K are con
#stants.
```

```python
import ROOT 
from ROOT import TH1F,TPaveText,TGraph,TF1


import numpy as np
from ROOT import TCanvas
```

```python
#correction factor for Tungsten 4.4 eta.

Correction_Factor=[1.371742112,1.205109665,1.158211721,1.143379831,1.136105431,1.125365744,1.11691909]
Incident_Energy = [100,250,500,750,1000,2000,10000]


correction_factor= np.asarray(Correction_Factor,dtype= np.float64)
incident_energy= np.asarray(Incident_Energy,dtype= np.float64)
entries=len(incident_energy)






correction_factor_graph=TGraph(entries,incident_energy, correction_factor)

fit_function_str = 'x/(x*(1-[0])-[1])'
fit_function = TF1('cf_fit',fit_function_str,min(incident_energy)*0.9,max(incident_energy)*1.1)

fit_result=correction_factor_graph.Fit(fit_function,'0 M S Q R')
    
fit_result.Print()

Canvas=TCanvas("","")
Canvas.Draw('')

fit_function.Draw('')

correction_factor_graph.Draw('*')



```
